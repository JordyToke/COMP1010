## Using python as a calculator (0.5 marks)

Write snippets of python code in the cells provided to calculate the answer to each of these questions. 

How many inches are there in 37 metres (approximately)?

````python
#If there are ~39.37inches in a metre 
metres = 37
conversion = 39.37
inches = metres * conversion
print(f"Approximately {inches:.0f} inches.")
````

Bill is *exactly* 31 years and 68 days old. How old is Bill in seconds (assuming a year is on average 365.25 days)?

````python
days = 1
years = 365.25*days
hours = 24
minutes = 60
seconds = 60
bill_in_days = 31*years + 68
bill_in_seconds = bill_in_days * hours * minutes * seconds
print(f"Bills age in seconds is {bill_in_seconds}")
````

## Conditionals (0.5 marks)

When visiting a theme park, many rides require you to be a certain height before you can ride them.

At a specific park there are:
* Red Riders: 85-105cm tall. They have access to 9 rides.
* Green Riders: 106-131cm tall. They have access to 21 rides.
* Yellow Riders: 132cm and taller. They have access to 27 rides.

Write a program which asks the user for their height and prints out how many rides they can go on.

Write this program in two different ways.

````python
# First approach
rider_height = int(input('Please enter your height in centimetres: '))
if rider_height < 85:
  print("Unfortunately you are not tall enough to ride.")
elif rider_height < 106:
  print('You may access 9 of the rides.')
elif rider_height < 132:
  print('You may access 21 of the rides.')
else:
  print('You may access 27 of the rides.')
````

````python
# Second approach
rider_height = int(input('Please enter your height in centimetres: '))
rides = ''
red = 9
green = 21
yellow = 27
if rider_height > 131:
  rides = yellow
elif rider_height > 105:
  rides = green
elif rider_height > 84:
  rides = red
else:
  rides = 0

print(f"You may use {rides} of the rides.")
````
